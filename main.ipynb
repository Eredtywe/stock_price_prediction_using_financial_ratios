```python
import numpy as np
import pandas as pd
import sklearn
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
data['D/E'] = data['liabilities'] / data['equity']
data['current_ratio'] = data['current_assets'] / data['current_liabilities']
data['profit_margin'] = data['net_income'] / data['total_revenue']
data['EPS'] = data['net_income'] / data['shares_outstanding']
```

```python
data = data.drop('Date', axis=1)
data = data.drop('Stock', axis=1)
data = data.drop('liabilities', axis=1)
data = data.drop('equity', axis=1)
data = data.drop('total_assets', axis=1)
data = data.drop('current_assets', axis=1)
data = data.drop('current_liabilities', axis=1)
data = data.drop('total_revenue', axis=1)
data = data.drop('net_income', axis=1)
data = data.drop('dividend', axis=1)
data = data.drop('shares_outstanding', axis=1)
```
